## Question 1.	Write a Pyspark code for decision tree like exercise you have completed in pyhton. 

```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
from pyspark.streaming import StreamingContext
```

```python
from pyspark.mllib.tree import DecisionTree, DecisionTreeModel
from pyspark.mllib.util import MLUtils
from pyspark.mllib.regression import LabeledPoint
from pyspark.ml.feature import StringIndexer
```

```python
sc = SparkContext.getOrCreate()
```

```python
spk = SparkSession.builder.getOrCreate()
```

```python
data = spk.read.csv("data.csv")
```

```python
data.show(n = 5)
```

```python
cols = data.columns
cols
```

```python
data.show(n=1000)
```

```python
# In RDD format:
rdd_data = data.rdd
```

```python
rdd_data.collect()
```

```python
#changing to LabeledPoint for modelling: 
rdd_data = rdd_data.map(lambda row: LabeledPoint(row[4], row[0:4]))
```

```python
rdd_data.collect()
```

```python
#Train and test:
(train, test) = rdd_data.randomSplit([0.7, 0.3])
```

```python
train.collect()
```

```python
test.collect()
```

```python
model_dt = DecisionTree.trainClassifier(train, numClasses=2, categoricalFeaturesInfo={}, impurity='gini', maxDepth=5, maxBins=32)
```

```python
# Model Evaluation on test instances and compute test error
predictions = model_dt.predict(test.map(lambda x: x.features))
labelsAndPredictions = test.map(lambda lp: lp.label).zip(predictions)
testErr = labelsAndPredictions.filter(
    lambda lp: lp[0] != lp[1]).count() / float(test.count())
print('Test Error = ' + str(testErr))
print('Learned classification tree model:')
print(model_dt.toDebugString())
```

## 4. c.	Find top occupation for every year.


```python
from collections import Counter
top_occup = rows.map(lambda x: (x[0],x[1])).groupByKey().mapValues(list).map(lambda x: (x[0],Counter(x[1]).most_common(1)))
top_occup.sortByKey().collect()
```

## 4. d.	Which guest have attended maximum shows
